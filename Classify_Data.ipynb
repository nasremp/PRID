```python
directory= r'/content/drive/MyDrive/MY PRID/detection output for the 6 cameras' # r indicates for raw string to avoid spaces in directories
counter=0
for camera in os.listdir(directory):
    c = os.path.join(directory, camera)
    print(camera) # get inside folder with 6 folders (camera1, camera2, camera3, camera4, camera5, camera6)
    for out in os.listdir(c):
      o=os.path.join(c, out) #get inside 'out' folder in each camera
      for PID in os.listdir(o):
        if counter%2 == 0:
          p=os.path.join(o,PID)
          print(PID) #get inside each folder of 'out', each folder named with a number that represents person ID for only one person with multiple images
          count = len(fnmatch.filter(os.listdir(p), '*.*')) #count number of images per person
          #print(count)
        counter=counter+1
        for image in os.listdir(p): #Uncomment this loop if you want to get each image for PID
            i=os.path.join(p, image) #get the path of each image
            image_number=int(''.join(filter(str.isdigit, image))) #get only the number of random image without any extensions
            image_new_name=PID+'_'+camera+'_'+str(image_number) #name this image with the same format of DukeMTMC-reID
            new_path='/content/drive/MyDrive/MY PRID/ZC_DATASET_PRID/bounding_box_train'+'/'+image_new_name+'.jpg'
            #print(i)
            #print(new_path)
            shutil.copyfile(i, new_path)
            
```
